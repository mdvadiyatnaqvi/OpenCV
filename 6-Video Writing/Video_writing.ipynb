### Video Writing

```python
# reading video file from source/computer

import cv2

source = 0 # give you video file path

cam = cv2.VideoCapture(source) # as we have 0 in source, so it'll use the camera for us
```

```python
ret, frame = cam.read()
```

```python
import matplotlib.pyplot as plt
plt.imshow(frame[..., ::-1])
```

### Writing video by using cv2

```python
width = int(cam.get(3))
height = int(cam.get(4))

out_avi = cv2.VideoWriter("demo.avi", cv2.VideoWriter_fourcc('M', 'J', 'P', 'G'), 60, (width, height))
```

```python
while cam.isOpened():
    ret, frame = cam.read()
    
    if ret == True:
        out_avi.write(frame)
    else:
        break
```

```python
cam.release()
out_avi.release()
cv2.destroyAllWindows()
```
