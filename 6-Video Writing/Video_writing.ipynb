### Video Writing

```python
# reading video file from source/computer

import cv2

source = 0 # give you video file path

cam = cv2.VideoCapture(source) # as we have 0 in source, so it'll use the camera for us
```

```python
ret, frame = cam.read()
```

```python
import matplotlib.pyplot as plt
plt.imshow(frame[..., ::-1])
```
