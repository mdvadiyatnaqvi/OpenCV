### Working with Color Images

```python
coka_img = cv2.imread('./Resources/Images/coka_cola.png')
print('Size of the Coka Cola Image is: ', coka_img.shape)
plt.imshow(coka_img)
plt.show()
```

```python
# converting into original image
original_coka_image = coka_img[:, :, ::-1]
plt.imshow(original_coka_image)
plt.show()
```

### Splitting and merging Channels b, g, r

```python
cimg = cv2.imread('./Resources/Images/coka_cola.png', cv2.IMREAD_COLOR)
b, g, r =cv2.split(cimg)
plt.figure(figsize=[20,5])
plt.subplot(141);plt.imshow(r, cmap='gray');plt.title('Red Channel')
plt.subplot(142);plt.imshow(g, cmap='gray');plt.title('Green Channel')
plt.subplot(143);plt.imshow(b, cmap='gray');plt.title('Blue Channel')
# mixing all channels
merged_img = cv2.merge((b, g, r))
plt.subplot(144);plt.imshow(merged_img[:, :, ::-1]);plt.title('Merged Image')
plt.show()
```

### Changing to hsv color space

```python
scape_img = cv2.imread('./Resources/Images/landscape.jpg')
# for conversion bgr to hsv(hue, saturation, value)
hsv_img=cv2.cvtColor(scape_img, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv_img) # for modification in color we can add value to any on channel like h+10
plt.figure(figsize=[20, 5])
plt.subplot(141);plt.imshow(h, cmap='gray');plt.title("H Channel")
plt.subplot(142);plt.imshow(s, cmap='gray');plt.title("S Channel")
plt.subplot(143);plt.imshow(v, cmap='gray');plt.title("V Channel")
plt.subplot(144);plt.imshow(scape_img);plt.title("Original")
plt.show()
```

### Saving images files to memory

```python
# imwrite function has been used to save the files.

scape_merged = cv2.merge((h, s, v))
scape_rgb = cv2.cvtColor(scape_merged, cv2.COLOR_HSV2RGB)
plt.imshow(scape_rgb)
plt.show()

# cv2.imwrite('./Resources/Saved Images/changed_landscape.jpg', hsv_img)
```
